Gaze Data Analysis

```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
file = 'gaze_data_19-03-15-46'
input_file = file + '.csv'
data = pd.read_csv(input_file)
# gaze data columns
# Select relevant columns for left and right gaze positions along the x-axis
left_gaze_x = data['LeftGazePoint2DX']
right_gaze_x = data['RightGazePoint2DX']

# Plot left gaze positions along the x-axis
plt.plot(left_gaze_x, label='Left Gaze X in normalized pixel coordinates')

# Plot right gaze positions along the x-axis
plt.plot(right_gaze_x, label='Right Gaze X in normalized pixel coordinates')

# Add labels and title
plt.xlabel('Timestamp')
plt.ylabel('Gaze Position (X-axis)')
plt.title('Gaze Positions along X-axis')
plt.legend()

# Show plot
plt.show()
```
