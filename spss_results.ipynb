### Mean and SEM plots of statistical results from SPSS

```python
import pandas as pd
import csv
from matplotlib import pyplot as plt

def get_XAI(idd):
    if 'X' in idd:
        return 1
    else:
        return 0

def get_spss_data(tracker_results):
    tracker_results['XAI_value'] = tracker_results['ID'].apply(get_XAI)
    tracker_results['elmo_fixation_rate'] = tracker_results['elmo_fixation_count']/tracker_results['game_time']
    tracker_results['computer_fixation_rate'] = tracker_results['computer_fixation_count']/tracker_results['game_time']

    tracker_results['elmo_fixation_percentage'] = tracker_results['elmo_total_fixation_duration']/(tracker_results['elmo_total_fixation_duration']+tracker_results['computer_total_fixation_duration'])
    tracker_results['computer_fixation_percentage'] = tracker_results['computer_total_fixation_duration']/(tracker_results['elmo_total_fixation_duration']+tracker_results['computer_total_fixation_duration'])

    tracker_results['elmo_probability_fixation'] = tracker_results['elmo_fixation_count']/(tracker_results['elmo_fixation_count']+tracker_results['computer_fixation_count'])
    tracker_results['computer_probability_fixation'] = tracker_results['computer_fixation_count']/(tracker_results['elmo_fixation_count']+tracker_results['computer_fixation_count'])
    
    tracker_results['elmo_probability_to_computer'] = tracker_results['elmo_fixation_count']/tracker_results['computer_fixation_count']
    tracker_results['elmo_proportion_to_computer'] = tracker_results['elmo_total_fixation_duration']/tracker_results['computer_total_fixation_duration']

    
    tracker_results.to_csv('tracker_results_final_XAI_numeric.csv', index=False)

tracker_results = pd.read_csv('tracker_results_final_XAI_numeric.csv')
get_spss_data(tracker_results)

elmo_count_o = 127.64
elmo_count_ox = 315
elmo_count_sem = 38.455
elmo_count_ox_sem = 38.455

computer_count_o = 399.43
computer_count_ox =  539
computer_count_sem =  36.509
computer_count_ox_sem = 60.024

positions = [0, 10]
labels = ['False', 'True']
y_index = [0, 10]
    
plt.figure(figsize=(10, 6))
plt.errorbar(y_index, [elmo_count_o, elmo_count_ox], [elmo_count_sem, elmo_count_ox_sem], color='black', marker = 'o', markeredgecolor = 'black', markerfacecolor = 'white', ls = '-', capsize=5, capthick=1, ecolor='black')
plt.errorbar(y_index, [computer_count_o, computer_count_ox], [computer_count_sem, computer_count_ox_sem], markeredgecolor = 'black', color='black', marker = 'o', ls = '--', capsize=5, capthick=1, ecolor='black')

# plt.plot(y_index, [elmo_count_o, elmo_count_ox], color='tab:blue', linestyle = '-')
# plt.plot(y_index, [computer_count_o, computer_count_ox], color='tab:orange', linestyle = '-')
handles = [
        plt.Line2D([], [], color='black', linestyle='-', label='Robot Elmo'),
        plt.Line2D([], [], color='black', linestyle='--', label='Computer'),
    ]
plt.legend(handles=handles)

plt.ylabel('Total Fixation Count')
plt.xlabel('XAI Model')
plt.xticks(positions, labels)
# plt.title('Elmo Fixation Count')
plt.savefig('fixation_count_means_plot.png', dpi = 1600, bbox_inches='tight')
plt.show()


```
